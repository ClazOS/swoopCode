```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from zipline.utils.factory import load_bars_from_yahoo
end = pd.Timestamp.utcnow()
start = end - 2500 * pd.tseries.offsets.BDay()
```

```python
data = load_bars_from_yahoo(stocks=['XIV', 'VXX', 'UVXY', 'TVIX',
  'ZIV','SPY','SPXL'],
  start=start, end=end)
```

```python
prices = data.minor_xs('price').dropna()
returns = np.log(prices / prices.shift(1)).dropna()
def compound(R):
  return (1 + R).cumprod()
```

```python
spreads = [
# -(returns.VXX + returns.XIV),
  -(returns.SPXL - 3*returns.SPY),
# -(returns.UVXY - 2*returns.VXX),
# -(returns.TVIX - 2*returns.VXX),
# -(returns.UVXY + 2*returns.XIV),
# -(returns.TVIX + 2*returns.XIV)
]
```

```python
yields = pd.DataFrame([compound(r) for r in spreads]).T
```

```python
yields.plot()
```

```python
weights = np.linspace(-0.9, -0.1, 25)
```

```python
portfolios = pd.DataFrame({
  weight: weight*returns.VXX - (1 + weight)*returns.XIV
  for weight in weights
})
```

```python
cumR = (1 + portfolios).cumprod()
cumR.plot(figsize=(12,8), legend=False)
```
