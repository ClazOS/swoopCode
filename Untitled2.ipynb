```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
from zipline.utils.factory import load_bars_from_yahoo
import pandas as pd
```

```python
def get_portfolio(fname=None):
   if fname is None:
       fname = "portfolio/portfolio.csv"
   base = os.path.expanduser("~/")
   path = base + "/dropbox/EllieMae/" + fname
   portfolio = pd.read_csv(path)
   return portfolio
```

```python
def get_portfolio_data(days=60, fname=None):
   port = get_portfolio(fname=fname)
   now = pd.Timestamp.utcnow()
   bdays = days*pd.tseries.offsets.BDay()
   data = load_bars_from_yahoo(
       stocks=port['symbol'],
       start=now - bdays,
       end=now)
   return data
```

```python
def port_plot(prices):
   rows = len(prices.columns)
   fig, axes = plt.subplots(rows, 1, figsize=(10,rows*3))
   for i in range(rows):
       series = prices[prices.columns[i]]
       axes[i].plot(series)
       axes[i].set_title(prices.columns[i])
   fig.tight_layout()

   
data = get_portfolio_data()
prices = data.minor_xs('price')
port_plot(prices)
```

```python
def normalize_prices(prices, window):
   return prices.iloc[-window:] / prices.iloc[-window]


def get_weights(fname=None):
   port = get_portfolio(fname=fname)
   return pd.Series(port['weight'].values, 
                    index=port['symbol'].values)

weights = get_weights()
weighted_port = prices * weights

normalize_prices(weighted_port, 30).plot(figsize=(12,8))
```
